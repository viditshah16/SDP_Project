```python
def getData(path):
    anger=0
    fear=0
    sad=0
    surprise=0
    happy=0
    neutral=0
    df=pd.read_csv(path)
    
    x=[]
    y=[]
    
    for i in range(len(df)):
        if df.iloc[i]['emotion']!=1:
            if df.iloc[i]['emotion']==0:
                if anger<=4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(x) for x in im.split()]
                    x.append(im)
                    anger+=1
                else:
                    pass
                
            if df.iloc[i]['emotion']==2:
                if fear<=4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(x) for x in im.split()]
                    x.append(im)
                    fear+=1
                else:
                    pass
                
            if df.iloc[i]['emotion']==3:
                if happy<=4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(x) for x in im.split()]
                    x.append(im)
                    happy+=1
                else:
                    pass
                
            if df.iloc[i]['emotion']==4:
                if sad<=4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(x) for x in im.split()]
                    x.append(im)
                    sad+=1
                else:
                    pass
                
            if df.iloc[i]['emotion']==5:
                if surprise<=4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(x) for x in im.split()]
                    x.append(im)
                    surprise+=1
                else:
                    pass
            
            if df.iloc[i]['emotion']==6:
                if neutral<=4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(x) for x in im.split()]
                    x.append(im)
                    neutral+=1
                else:
                    pass
        
    return x ,y
                
                
                
                
                
                
                
                    
```

```python
x , y = getData('../input/facial-expression/fer2013.csv')
```

```python
np.unique(y , return_counts=True)
```

```python
x=np.array(x)/255.0
y=np.array(y)
```

```python
x.shape , y.shape
```

```python
y_o = []
for i in y:
    if i!=6:
        y_o.append(i)
    else:
        y_o.append(1)
```

```python
np.unique(y_o , return_counts=True)
```

```python
x = x.reshape(len(x) , 48 , 48 , 1)
#no.of images , height , width , color_map 
```

```python
x.shape
```

```python
import keras
from tensorflow.keras.utils import to_categorical
y_new = to_categorical(y_o , num_classes = 6)
```

```python
len(y_o) , y_new.shape
```

```python
y_o[0] , y_new[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation, Dropout, Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.metrics import categorical_accuracy
from keras.models import model_from_json
from keras.callbacks import ModelCheckpoint
from keras.optimizers import *
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.optimizers import Adam
```

```python
emotion_model = Sequential()

emotion_model.add(Conv2D(64,(5, 5), activation='relu', input_shape=(48,48,1) , padding='same'))
emotion_model.add(Conv2D(64, (5, 5), padding='same'))
emotion_model.add(BatchNormalization())
emotion_model.add(Activation('relu'))
emotion_model.add(MaxPooling2D(pool_size=(2, 2)))

emotion_model.add(Conv2D(128, (5, 5), activation='relu',padding='same'))
emotion_model.add(Conv2D(128, (5, 5), padding='same'))
emotion_model.add(BatchNormalization())
emotion_model.add(Activation('relu'))
emotion_model.add(MaxPooling2D(pool_size=(2, 2)))

emotion_model.add(Conv2D(256, (3, 3), activation='relu',padding='same'))
emotion_model.add(Conv2D(256, (3, 3), activation='relu',padding='same'))
emotion_model.add(BatchNormalization())
emotion_model.add(MaxPooling2D(pool_size=(2, 2)))

emotion_model.add(Flatten())
emotion_model.add(Dense(6, activation='softmax'))

emotion_model.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.0001, decay=1e-6),metrics=['accuracy'])
```

```python
emotion_model.fit(x , y_new, epochs=22 , batch_size=64 , shuffle=True , validation_split=0.2)
```

```python
emotion_model.save('emotion_model.h5')
```
